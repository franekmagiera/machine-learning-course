In two above figures there is a noticable linear correlation between price of a house and its square footage and number of bedrooms. I will use a linear regression model to predict house prices. Hypothesis function is given as
$h(x) = \theta^T  x$,where theta is a vector of line's coefficients and x is a vector of features(x1 - area in square feet, x2 - number of bedrooms). I have to introduce additional feature x0 equal to 1 so $h(x) = \theta_0  x_0 + \theta_1  x_1 + \theta_2  x_2 = \theta_0 + \theta_1  x_1 + \theta_2  x_2$ in order to get a line that does not have to go through $(0,0)$ for $(x_1,x_2)$. Hypothesis function returns predicted house price depending on the input feature vector x. The goal is to find coefficients theta so that the mean square error between h(x) and y (y - vector of house prices in the dataset) for every entry in the provided data set is minimal: $$J(\theta) = \frac{1}{2m}\sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})^2 \rightarrow min$$ where m is the number of entries in the data set, $x^{(i)}$ is a vector of features of the i-th entry and $y^{(1)}$ corresponds to the price of the house of the i-th entry. Because $J(\theta)$ is a convex function, it only has one global minimum and no local minima, hence gradient descent algorithm is a very good method of finding this global minimum (its result does not depend on the starting point in this instance).

Because features in the data set differ by 3 orders of magnitude I have to perform feature normalization prior so my implementation of gradient descent will run faster.

```python
normalized_data = (data.drop('Price in dollars', axis=1) - data.drop('Price in dollars', axis=1).mean())/data.drop('Price in dollars', axis=1).std()
normalized_data = pd.DataFrame({'x0' : [1 for i in range(m)]}).join(normalized_data).join(data['Price in dollars'])
normalized_data.head()
```

Cost function $J(\theta)$ can also be expressed as $\frac{1}{2m}(X\theta - y)^T (X\theta - y)$:

```python
def J(theta, X, y):
    '''theta is a vector of coefficients, X is a data set, y is a vector of house prices'''
    
    m = len(X)
    return 1/(2*m) * (X.dot(theta)-y).T.dot(X.dot(theta)-y)

X = normalized_data[['x0', 'Area in square feet', 'Number of bedrooms']].values
y = normalized_data[['Price in dollars']].values
starting_theta = np.zeros((3,1))
J(starting_theta, X, y)
```

By computing the derivative of $J(\theta)$ we get the formula for $\theta_j$ coefficients. $$\theta_j := \theta_j - \frac{\alpha}{m}\sum_{i}^{m} (h(x^{(i)}) - y^{(i)})x^{(i)}_j$$ where $\alpha$ is the learning rate. Gradient descent:

```python
def gradient_descent(theta, X, y, alpha, iterations):
    '''This function performs gradient descent for J(theta)'''
    
    m = len(X)
    # vector preallocation
    J_history = np.zeros(iterations+1) # vector containing information on how cost function changes during gradient descent
    h = np.zeros(m) # hypothesis function values
    J_history[0] = J(theta, X, y)
    for i in range(iterations):
        h = np.array([theta.T.dot(X[j]) for j in range(m)])
        theta = np.array([theta[j] - alpha/m * np.sum(np.multiply(np.subtract(h, y), np.reshape(X.T[j], (m,1)))) for j in range(len(theta))])
        
        J_history[i+1] = J(theta, X, y)
    
    return np.array([J_history, theta])

iterations = 50
result1 = gradient_descent(starting_theta, X, y, 0.03, iterations)
result2 = gradient_descent(starting_theta, X, y, 0.1, iterations)
result3 = gradient_descent(starting_theta, X, y, 0.3, iterations)
plt.plot(np.arange(iterations+1), result1[0], label = r'$\alpha=0.03$')
plt.plot(np.arange(iterations+1), result2[0], label = r'$\alpha=0.1$')
plt.plot(np.arange(iterations+1), result3[0], label = r'$\alpha=0.3$')
plt.title('Minimization of the cost function with gradient descent')
plt.xlabel('Iterations')
plt.ylabel(r'J($\theta$)')
plt.tight_layout()
plt.legend()
```

On the above graph one can notice the impact that choosing the learnig rate $\alpha$ has on gradient descent method convergence. If it is too small algorithm will converge very slowly. On the other hand if it is too large it will not converge and may even diverge.

Now for example I can predict how much a 1650 square foot house with 3 bedrooms should be. But first I have to normalize the features before using the hypothesis function I got with gradient descent.

```python
theta = result2[1]
x0 = 1
x1 = 1650
x2 = 3
x1 = (x1 - data['Area in square feet'].mean())/data['Area in square feet'].std()
x2 = (x2 - data['Number of bedrooms'].mean())/data['Number of bedrooms'].std()
x = np.array([x0, x1, x2]).reshape(3,1)
predicted_price = theta.T.dot(x)[0][0]
print("Predicted price of 1650 square foot house with 3 bedrooms is ${:.2f}".format(predicted_price))
```

There is also another method that does not require feature scaling - using normal equations. The $\theta = (X^T X)^{-1} X^T y$ formula is the result of differentiating the cost function and using some clever transformations.

```python
data2 = pd.DataFrame({'x0' : [1 for i in range(m)]}).join(data)
X = data2[['x0', 'Area in square feet', 'Number of bedrooms']].values
y = data2[['Price in dollars']].values
theta = (np.linalg.inv(X.T.dot(X))).dot(X.T).dot(y)
x0 = 1
x1 = 1650
x2 = 3
x = np.array([x0, x1, x2]).reshape(3,1)
predicted_price2 = theta.T.dot(x)[0][0]
print("Predicted price of 1650 square foot house with 3 bedrooms is ${:.2f}.\nThe difference between normal equation result and gradient descent result is {:.2f}".format(predicted_price2, predicted_price2 - predicted_price))
```

Both methods give roughly the same result. For small problems like this using normal equations is better, because it is faster to write and calculations are performed quite fast. For very large problems where inverting large matrices will take a lot of time it would be better to use gradient descent.
