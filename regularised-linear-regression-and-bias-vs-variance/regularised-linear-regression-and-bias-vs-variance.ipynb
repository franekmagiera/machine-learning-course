The goal of this exercise is to implement regularised linear regression and use it to study models with different bias and variance properties. I have a data set provided by the course lecturer containing records of the changes of the water level in a dam and the amount of water flowing out of the dam. The goal is to predict how much water is flowing out of the dam given the changes of water level. I will start by importing necessary libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

The dataset is divided into a training set, a cross validation set and a test set. Training set will be used to train the model (learn linear regression parameters), cross validation set will be used to determine the regularisation parameter and test set will be used to test the performance of the model on previously unseen, new examples.

```python
X = np.genfromtxt('X.csv', delimiter=';')
Xval = np.genfromtxt('Xval.csv', delimiter=';')
Xtest = np.genfromtxt('Xtest.csv', delimiter=';')
y = np.genfromtxt('y.csv', delimiter=';')
yval = np.genfromtxt('yval.csv', delimiter=';')
ytest = np.genfromtxt('ytest.csv', delimiter=';')
```

Visualisation of the training set:

```python
plt.figure(figsize=(16/1.5, 10/1.5))
plt.plot(X, y, 'rx')
plt.xlabel('Change in water level')
plt.ylabel('Water flowing out of the dam')
plt.title('Training set');
```

Regularised linear regression cost function is: $J(\theta) = \frac{1}{2m} (\sum_{i=1}^m (h_\theta(x^{(i)}) - y^{(i)})^2) + \frac{\lambda}{2m} (\sum_{j=1}^n \theta_j^2)$.

```python
def lin_reg_cost_function(theta, X, y, regularisation_parameter):
    m = X.shape[0]
    n = theta.shape[0]
    if len(X.shape) == 1:
        X = np.column_stack((np.ones((m,1)), X))
    else:
        X = np.insert(X, 0, 1, axis=1)
    H_theta = X @ theta
    J = 1/(2*m) * np.sum(np.sum(np.square(np.subtract(H_theta.T, y))))
    J += regularisation_parameter/(2*m) * np.sum(np.sum(np.square(theta[1:])))
    grad = np.zeros((theta.shape[0], 1))
    for j in range(len(grad)):
        grad[j] = 1/m * np.sum(np.sum(np.multiply(np.subtract(H_theta.T, y), X[:, j])))
    np.multiply(regularisation_parameter/m, theta[1:])
    grad[1:] += np.multiply(regularisation_parameter/m, theta[1:].reshape((theta[1:].shape[0], 1)))
    return [J, grad.ravel()]
theta = np.ones((2, 1))
lin_reg_cost_function(theta, X, y, 0)
```

Now it is time to train the model. I will not use regularisation because $\theta$ is 2 dimensional and it would not be very helpful.

```python
from scipy import optimize
opt_theta = optimize.fmin_cg(f=lambda x: lin_reg_cost_function(x, X, y, 0)[0],
                             x0=np.ones(2),
                             fprime=lambda x: lin_reg_cost_function(x, X, y, 0)[1])
```

```python
plt.figure(figsize=(16/1.5, 10/1.5))
plt.plot(X, y, 'rx')
plt.plot(X, (np.column_stack((np.ones((X.shape[0], 1)), X))@opt_theta).ravel())
plt.xlabel('Change in water level')
plt.ylabel('Water flowing out of the dam')
plt.title('Training set');
```

The clou of the whole exercise is to learn about the balance between high bias and high variance. Models with high bias are often too simple and tend to underfit the data while models with high variance are very complex and are overfitting the data. A good way to determine whether the model suffers from either of those is to plot and analyse the learning curves. A learning curve is a function of a training set size and it returns a cost function's value. I will compute training errors only for training subsets whereas cross validation error will be computed for the whole cross validation set. I will not include regularisation errors in the learning curves.

```python
# Learning curves
def plot_learning_curves(X, Xval, y, yval, regularisation_parameter):
    m = X.shape[0]
    errors_train = np.zeros(m)
    errors_val = np.zeros(m)
    for i in range(m):
        if len(X.shape) == 1: # one dimension
            training_subset = X[:i+1]
            n = 2
        else:
            training_subset = X[:i+1, :]
            n = X.shape[1] + 1
        y_subset = y[:i+1]
        opt_theta = optimize.fmin_cg(f=lambda x: lin_reg_cost_function(x, training_subset, y_subset, regularisation_parameter)[0],
                                     x0=np.ones(n),
                                     fprime=lambda x: lin_reg_cost_function(x, training_subset, y_subset, regularisation_parameter)[1],
                                     disp=False)
        # Not taking regularisation into account for calculating error.
        errors_train[i] = lin_reg_cost_function(opt_theta, training_subset, y_subset, 0)[0]
        errors_val[i] = lin_reg_cost_function(opt_theta, Xval, yval, 0)[0]
    plt.figure(figsize=(16/1.5, 10/1.5))
    plt.plot(range(m), errors_train, label='Training')
    plt.plot(range(m), errors_val, label='Cross validation')
    plt.title('Learning curves for linear regression')
    plt.xlabel('Training dataset size')
    plt.ylabel('Cost function value')
    plt.legend()
plot_learning_curves(X, Xval, y, yval, 0)
```

Above learnig curves are high both for cross validation set and training set. This is indicative of high bias - the model is to simple and cannot fit the dataset well. To remedy this I will use more polynomial features so the new hypothesis has the following form: $h_\theta(x) = \theta_0 + \theta_1 (water level) + \theta_2 (water level)^2 + ... + \theta_p (water level)^p = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + ... + \theta_p x_p$. I will now implement a function that given the original training set maps it into a bigger with polynomial features.

```python
def poly_features(X, p):
    # Preallocation.
    X_poly = np.ones((len(X), p))
    for i in range(p):
        X_poly[:, i] = np.power(X, i+1)
    return X_poly

X_poly = poly_features(X, 8)
X_poly_val = poly_features(Xval, 8)
```

Features now differ from each other by several orders of magnitude. The learning algorithm will not work very well. Performing feature normalisation is a good preprocessing step.

```python
def normalise_features(X):
    means = np.vstack((np.mean(X, axis=0),) * X.shape[0])
    stds = np.vstack((np.std(X_poly, axis=0),) * X.shape[0])
    X_norm = np.divide(np.subtract(X, means), stds)
    return X_norm

X_norm = normalise_features(X_poly)
X_norm_val = normalise_features(X_poly_val)
```

Now I will see how this model performs and plot learnig curves. I will not use regularisation at first.

```python
plot_learning_curves(X_norm, X_norm_val, y, yval, 0)
```

The training error is equal to 0 regardless of the training set size. However cross validation error is quite high. This means that the model adapted very well to the training data but it does not generalise well. One way to remedy this is to use regularisation.

```python
plot_learning_curves(X_norm, X_norm_val, y, yval, 1)
```

```python
plot_learning_curves(X_norm, X_norm_val, y, yval, 100)
```

We can see that using regularisation parameter equal to 1 we got much better results on a cross validation set. However the parameter cannot be to big because then again it does not fit the data well. To choose a good value of a regularisation parameter I will calculate the training and cross validation errors depending on the regularistaion parameter's value and plot them to choose the parameter that gives the smallest cross validation error.

```python
# Learning curves
def plot_error_reg(X, Xval, y, yval, regularisation_parameters):
    m = X.shape[0]
    errors_train = np.zeros(len(regularisation_parameters))
    errors_val = np.zeros(len(regularisation_parameters))
    for i in range(len(regularisation_parameters)):
        if len(X.shape) == 1: # one dimension
            n = 2
        else:
            n = X.shape[1] + 1
        opt_theta = optimize.fmin_cg(f=lambda x: lin_reg_cost_function(x, X, y, regularisation_parameters[i])[0],
                                     x0=np.ones(n),
                                     fprime=lambda x: lin_reg_cost_function(x, X, y, regularisation_parameters[i])[1],
                                     disp=False)
        # Not taking regularisation into account for calculating error.
        errors_train[i] = lin_reg_cost_function(opt_theta, X, y, 0)[0]
        errors_val[i] = lin_reg_cost_function(opt_theta, Xval, yval, 0)[0]
    plt.figure(figsize=(16/1.5, 10/1.5))
    plt.plot(regularisation_parameters, errors_train, label='Training')
    plt.plot(regularisation_parameters, errors_val, label='Cross validation')
    plt.title('Learning curves for linear regression')
    plt.xlabel('Regularisation parameter')
    plt.ylabel('Cost function value')
    plt.ylim([0, 20])
    plt.legend()
plot_error_reg(X_norm, X_norm_val, y, yval, [0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10])
```

From the above plot we can see that the lowest cross validation error was achieved for regularisation parameter equal to 3. Now it is time to test model's performance on a test set containing examples that were neither used during learning linear regression parameters nor while choosing the regularisation parameter.

```python
final_theta = optimize.fmin_cg(f=lambda x: lin_reg_cost_function(x, X_norm, y, 3)[0],
                               x0=np.ones(9),
                               fprime=lambda x: lin_reg_cost_function(x, X_norm, y, 3)[1],
                               disp=False)
X_poly_test = poly_features(Xtest, 8)
X_norm_test = normalise_features(X_poly_test)
final_error = lin_reg_cost_function(final_theta, X_norm_test, ytest, 0)[0]
final_error
```

The test error is around 5.57.
